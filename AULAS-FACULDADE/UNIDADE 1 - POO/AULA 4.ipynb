# **Matplotlib com Pandas**

```python
import pandas as pd
import matplotlib.pyplot as plt

#Criando um DateFrame de exemplo
data = {'Ano': [1982, 1974, 2002, 2006, 2011, 2014, 2025], 'Vendas': [10, 10, 20, 30, 40, 50, 100]}

df =pd.DataFrame(data)

#Plotando um Gráfico de Linha

plt.figure(fisize=(12,8))
plt.plot(df['Ano'], df['Vendas'], marker='o', color='b', label='vendas')
```
