# **Matplotlib com Pandas**

```python
import pandas as pd
import matplotlib.pyplot as plt

#Criando um DateFrame de exemplo
data = {'Ano': [1974, 1982, 2002, 2006, 2011, 2014, 2025], 'Vendas': [10, 11.5, 20, 30, 40, 50, 100]}

df =pd.DataFrame(data)

#Plotando um Gráfico de Linha

plt.figure(figsize=(8,4))
plt.plot(df['Ano'], df['Vendas'], marker='o', color='b', label='vendas')
plt.title('Vendas ao Longo dos Anos')
plt.xlabel('Ano')
plt.ylabel('Vendas')
plt.grid(True)
plt.legend()
plt.show()
```

## ✅ Como transformar o gráfico de linha em um histograma (gráfico de barras)

🚫 Gráfico de linha:

    plt.plot(df['Ano'], df['Vendas'], marker='o', color='b', label='vendas')

✅ Substituir por gráfico de barras (histograma):

    plt.bar(df['Ano'], df['Vendas'], color='b', label='vendas')

📌 Diferença:

plt.plot() → usado para gráficos de linha (útil para séries temporais).

plt.bar() → usado para gráficos de barras (ideal para comparações diretas entre categorias).

```python
import pandas as pd
import matplotlib.pyplot as plt

#Criando um DateFrame de exemplo
data = {'Ano': [1974, 1982, 2002, 2006, 2011, 2014, 2025], 'Vendas': [10, 11.5, 20, 30, 40, 50, 100]}

df =pd.DataFrame(data)

#Plotando um Gráfico de Linha

plt.figure(figsize=(8,4))
plt.bar(df['Ano'], df['Vendas'], color='b', label='vendas')
plt.title('Vendas ao Longo dos Anos')
plt.xlabel('Ano')
plt.ylabel('Vendas')
plt.grid(True)
plt.legend()
plt.show()
```

# **Biblioteca - NumPy**
## **Integração com a Pandas**

```python
import pandas as pd
import numpy as np

# Criando um DataFrame de exemplo
data = {'A': [1, 2, 3, 4, 5], 'B': [10, 20, 30, 40, 50]}
df = pd.DataFrame(data)

# Usando o NumPy para calcular a média dos valores da coluna 'A' 
media = np.mean(df['A'])
print("Média da coluna 'A':", media)

# Usando o NumPy para calcular o desvio padrão dos valores da coluna 'B'
desvio_padrao = np.std(df['B'])
print("Desvio padrão da coluna 'B':", desvio_padrao)
```
