# **Matplotlib com Pandas**

```python
import pandas as pd
import matplotlib.pyplot as plt

#Criando um DateFrame de exemplo
data = {'Ano': [1974, 1982, 2002, 2006, 2011, 2014, 2025], 'Vendas': [10, 11.5, 20, 30, 40, 50, 100]}

df =pd.DataFrame(data)

#Plotando um GrÃ¡fico de Linha

plt.figure(figsize=(8,4))
plt.plot(df['Ano'], df['Vendas'], marker='o', color='b', label='vendas')
plt.title('Vendas ao Longo dos Anos')
plt.xlabel('Ano')
plt.ylabel('Vendas')
plt.grid(True)
plt.legend()
plt.show()
```

## âœ… Como transformar o grÃ¡fico de linha em um histograma (grÃ¡fico de barras)

ðŸš« GrÃ¡fico de linha:

    plt.plot(df['Ano'], df['Vendas'], marker='o', color='b', label='vendas')

âœ… Substituir por grÃ¡fico de barras (histograma):

    plt.bar(df['Ano'], df['Vendas'], color='b', label='vendas')

ðŸ“Œ DiferenÃ§a:

plt.plot() â†’ usado para grÃ¡ficos de linha (Ãºtil para sÃ©ries temporais).

plt.bar() â†’ usado para grÃ¡ficos de barras (ideal para comparaÃ§Ãµes diretas entre categorias).

```python
import pandas as pd
import matplotlib.pyplot as plt

#Criando um DateFrame de exemplo
data = {'Ano': [1974, 1982, 2002, 2006, 2011, 2014, 2025], 'Vendas': [10, 11.5, 20, 30, 40, 50, 100]}

df =pd.DataFrame(data)

#Plotando um GrÃ¡fico de Linha

plt.figure(figsize=(8,4))
plt.bar(df['Ano'], df['Vendas'], color='b', label='vendas')
plt.title('Vendas ao Longo dos Anos')
plt.xlabel('Ano')
plt.ylabel('Vendas')
plt.grid(True)
plt.legend()
plt.show()
```

# **Biblioteca - NumPy**
## **IntegraÃ§Ã£o com a Pandas**

```python
import pandas as pd
import numpy as np

# Criando um DataFrame de exemplo
data = {'A': [1, 2, 3, 4, 5], 'B': [10, 20, 30, 40, 50]}
df = pd.DataFrame(data)

# Usando o NumPy para calcular a mÃ©dia dos valores da coluna 'A' 
media = np.mean(df['A'])
print("MÃ©dia da coluna 'A':", media)

# Usando o NumPy para calcular o desvio padrÃ£o dos valores da coluna 'B'
desvio_padrao = np.std(df['B'])
print("Desvio padrÃ£o da coluna 'B':", desvio_padrao)
```
